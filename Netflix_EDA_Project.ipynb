# 5. Data Cleaning

```python
# 1. Initialize data_clean for data cleaning operations
df_clean = data.copy()

# Remove duplicates
before = len(df_clean)
df_clean.drop_duplicates(inplace=True)
print(f"Duplicates removed: {before - len(df_clean)}")
```

```python
# 2. Fill missing 'director' with 'Unknown'
#    Reason: dropping rows would lose 30% of data.
#    'Unknown' is a valid category for analysis.
df_clean['director'].fillna('Unknown', inplace=True)
```

```python
# 3. Fill missing 'country' with 'Unknown'
#    Reason: same as director — too many missing to drop.
df_clean['country'].fillna('Unknown', inplace=True)
```

```python
# 4. Drop rows where 'date_added' is missing
#    Reason: only 10 rows affected; we need this for time analysis.
df_clean.dropna(subset=['date_added'], inplace=True)
```

```python
# 5. Convert 'date_added' to datetime
#    Reason: stored as string; need datetime for year/month extraction.

# Ensure 'date_added' is string type before stripping, to handle re-execution
df_clean['date_added'] = df_clean['date_added'].astype(str)

df_clean['date_added'] = pd.to_datetime(
    df_clean['date_added'].str.strip(), errors='coerce', format='mixed'
)
df_clean['year_added'] = df_clean['date_added'].dt.year
df_clean['month_added'] = df_clean['date_added'].dt.month
```

```python
# 6. Drop irrelevant columns
#    Reason: 'cast' has 9.5% missing and won't be used in analysis.

# Check if 'cast' column exists before dropping
if 'cast' in df_clean.columns:
    df_clean.drop(columns=['cast'], inplace=True)

print(f"\nRows before cleaning: {len(data)}")
print(f"Rows after cleaning:  {len(df_clean)}")
print(f"\nRemaining missing values:")
print(df_clean.isnull().sum()[df_clean.isnull().sum() > 0])
print("\nData types after cleaning:")
print(df_clean.dtypes)
```

```python
df_clean['rating'].fillna(df_clean['rating'].mode()[0], inplace=True)
```

```python
# Drop rows where 'duration' is missing
initial_rows = len(df_clean)
df_clean.dropna(subset=['duration'], inplace=True)
print(f"Rows dropped due to missing duration: {initial_rows - len(df_clean)}")

# Verify no more missing 'duration' values
print("\nRemaining missing values after handling duration:")
print(df_clean.isnull().sum()[df_clean.isnull().sum() > 0])
```

```python
# 7. Verify cleaning
print("Shape after cleaning:", df_clean.shape)
print("Missing values after cleaning:\n", df_clean.isnull().sum())
print("Data types after cleaning:\n", df_clean.dtypes)
```

- Duplicate rows were removed to avoid repeated records.
- Missing values in director, cast, and country columns were replaced with meaningful labels.
- Missing ratings were filled using the mode value because it represents the most common category.
- The date_added column was converted into datetime format for easier analysis.
- After cleaning, the dataset became more consistent and suitable for exploratory data analysis.

# 6. Exploratory Data Analysis

##### Q1: Content type distribution?


```python
print("Q1: Content type distribution")
print("=" * 40)
type_counts = df_clean['type'].value_counts()
type_pct = df_clean['type'].value_counts(normalize=True) * 100
print(type_counts)
print("\nPercentage:")
print(type_pct.round(2))
```

##### Q2: Top 10 content-producing countries?


```python
print("\nQ2: Top 10 content-producing countries")
print("=" * 40)
top_countries = (df_clean[df_clean['country'] != 'Unknown']
                 ['country']
                 .str.split(',')
                 .explode()
                 .str.strip()
                 .value_counts()
                 .head(10))
print(top_countries)
```

##### Q3: Which month sees the most content additions?


```python
print("\nQ3: Which month sees the most content additions?")
print("=" * 40)
df_clean['month_added'] = df_clean['date_added'].dt.month_name()
df_clean['month_added'].value_counts()
```

##### Q4: Content added per year?


```python
print("\nQ4: Content added per year")
print("=" * 40)
yearly = df_clean.groupby('year_added').size().reset_index(name='count')
print(yearly.sort_values('year_added'))
```

##### Q5: Which year has the highest number of releases?

# 7. Visualizations

```python
plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
```

### Bar Chart

```python
# Movies vs TV Shows
plt.figure(figsize=(8,5))
sns.countplot(x='type', data=df_clean)
plt.title('Movies vs TV Shows on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

- Observation: Movies are more numerous than TV Shows.

### Line Chart

```python
#  Line Chart – Content release trend over years
plt.figure(figsize=(10,6))
data['release_year'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title("Content Release Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.show()
```

- Observation: Netflix content increased rapidly in recent years.

### Histogram

```python
# Distribution of release years

plt.figure(figsize=(10,6))
plt.hist(data['release_year'], bins=20, color='orange', edgecolor='black')
plt.title("Distribution of Release Years")
plt.xlabel("Release Year")
plt.ylabel("Frequency")
plt.show()
```

- Observation: Most titles belong to recent years.

### Scatter Plot

```python
# Release Year vs Duration

# First convert duration:
df_movies = df_clean[df_clean['type']=='Movie'].copy()

df_movies['duration_num'] = (
    df_movies['duration']
    .str.replace(' min','')
    .astype(float)
)

# Now plot:
plt.figure(figsize=(10,6))
plt.scatter(df_movies['release_year'], df_movies['duration_num'])
plt.title('Release Year vs Movie Duration')
plt.xlabel('Release Year')
plt.ylabel('Duration (Minutes)')
plt.show()
```

- Observation: Movie durations mostly fall within a specific range.

### Pie Chart

```python
# Content type distribution,
# Percentage of Movies and TV Shows
content_counts = df_clean['type'].value_counts()

plt.figure(figsize=(7,7))
plt.pie(content_counts, labels=content_counts.index, autopct='%1.1f%%')
plt.title('Percentage Distribution of Content Type')
plt.show()
```

- Observation: Movies occupy the largest share.

### Heatmap

```python
# # Correlation matrix
numeric_df = df_movies[['release_year','duration_num']]

plt.figure(figsize=(6,5))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

- Observation: Relationship between variables can be identified.

### Horizontal Bar

```python
# Top 10 countries by content count
top_countries = df_clean['country'].value_counts().head(10)

plt.figure(figsize=(10,6))
top_countries.plot(kind='barh', color='skyblue')
plt.title('Top 10 Countries by Content Count')
plt.xlabel('Number of Shows')
plt.show()
```

- Observation: The USA dominates Netflix content production.

```python
# Top Ratings Count
plt.figure(figsize=(10,6))
sns.countplot(y='rating', data=df_clean, order=df_clean['rating'].value_counts().index)
plt.title('Content Ratings Distribution')
plt.show()
```

- Observation: Certain rating categories appear much more frequently.

# 8. Insights Report

### 1.Netfix's catalog is movie-dominated.
- From the bar chart, movies significantly outnumber TV shows. This indicates Netflix’s primary focus remains on films, though TV shows are steadily growing
- Movies make up 69.7% of the library (6,128 titles) versus 30.3% TV Shows (2,666 titles), as shown in the bar and pie charts. This nearly 2:1 ratio suggests Netflix's acquisition and production strategy has historically prioritized movies, likely because films are cheaper to license in bulk than multi-season series.

### 2.Content addition grew explosively from 2016 to 2019, then declined.
- The line chart 0f yearly additions shows a steep climb from just 24 titles added in 2014 to a peak of 2,016 in 2019, followed by a drop to 1,879 (2020) and 1,498 (2021). This pattern suggests Netflix's aggressive content expansion phase peaked in 2019, after which the platform may have shifted toward quality/original content over volume — possibly influenced by the pandemic disrupting content pipelines.
- The yearly trend analysis showed a sharp increase in content additions after 2015. This reflects Netflix's aggressive global expansion and increasing investment in original and licensed content.

### 3.Netflix's library skews heavily toward recent content.
- The histogram of release years shows most titles cluster after 2010, with 2018 (1,146 titles), 2019 (1,030), and 2017 (1,031) being the top three release years. This indicates Netflix's catalog is built primarily around contemporary content rather than classic films and shows.

### 4.Movie durations are consist.
- The scatter plot of release year vs. duration reveals that most movies fall within a stable duration range (90–120 minutes). This suggests Netflix adheres to industry norms for viewer engagement.

### 5. Correlations among numeric features.
- The heatmap shows moderate correlations between release year and other numeric attributes. While not strongly predictive, these relationships can guide further modeling or recommendation systems.
