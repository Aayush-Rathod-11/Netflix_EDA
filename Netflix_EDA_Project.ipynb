# 5. Data Cleaning

```python
# Initialize data_clean for data cleaning operations
df_clean = data.copy()

# Remove duplicates
before = len(df_clean)
df_clean.drop_duplicates(inplace=True)
print(f"Duplicates removed: {before - len(df_clean)}")
```

```python
# 2. Fill missing 'director' with 'Unknown'
#    Reason: dropping rows would lose 30% of data.
#    'Unknown' is a valid category for analysis.
df_clean['director'].fillna('Unknown', inplace=True)
```

```python
# 3. Fill missing 'country' with 'Unknown'
#    Reason: same as director — too many missing to drop.
df_clean['country'].fillna('Unknown', inplace=True)
```

```python
# 4. Drop rows where 'date_added' is missing
#    Reason: only 10 rows affected; we need this for time analysis.
df_clean.dropna(subset=['date_added'], inplace=True)
```

```python
# 5. Convert 'date_added' to datetime
#    Reason: stored as string; need datetime for year/month extraction.

# Ensure 'date_added' is string type before stripping, to handle re-execution
df_clean['date_added'] = df_clean['date_added'].astype(str)

df_clean['date_added'] = pd.to_datetime(
    df_clean['date_added'].str.strip(), errors='coerce', format='mixed'
)
df_clean['year_added'] = df_clean['date_added'].dt.year
df_clean['month_added'] = df_clean['date_added'].dt.month
```

```python
# 6. Drop irrelevant columns
#    Reason: 'cast' has 9.5% missing and won't be used in analysis.

# Check if 'cast' column exists before dropping
if 'cast' in df_clean.columns:
    df_clean.drop(columns=['cast'], inplace=True)

print(f"\nRows before cleaning: {len(data)}")
print(f"Rows after cleaning:  {len(df_clean)}")
print(f"\nRemaining missing values:")
print(df_clean.isnull().sum()[df_clean.isnull().sum() > 0])
print("\nData types after cleaning:")
print(df_clean.dtypes)
```

```python
df_clean['rating'].fillna(df_clean['rating'].mode()[0], inplace=True)
```

```python
# Drop rows where 'duration' is missing
initial_rows = len(df_clean)
df_clean.dropna(subset=['duration'], inplace=True)
print(f"Rows dropped due to missing duration: {initial_rows - len(df_clean)}")

# Verify no more missing 'duration' values
print("\nRemaining missing values after handling duration:")
print(df_clean.isnull().sum()[df_clean.isnull().sum() > 0])
```

```python
# 7. Verify cleaning
print("Shape after cleaning:", df_clean.shape)
print("Missing values after cleaning:\n", df_clean.isnull().sum())
print("Data types after cleaning:\n", df_clean.dtypes)
```

- Duplicate rows were removed to avoid repeated records.
- Missing values in director, cast, and country columns were replaced with meaningful labels.
- Missing ratings were filled using the mode value because it represents the most common category.
- The date_added column was converted into datetime format for easier analysis.
- After cleaning, the dataset became more consistent and suitable for exploratory data analysis.

# 6. Exploratory Data Analysis

##### Q1: Content type distribution
##### Q2: Top 10 content-producing countries
##### Q3: Which month sees the most content additions?
##### Q4: Content added per year
##### Q5: Which year has the highest number of releases?

```python
print("Q1: Content type distribution")
print("=" * 40)
type_counts = df_clean['type'].value_counts()
type_pct = df_clean['type'].value_counts(normalize=True) * 100
print(type_counts)
print("\nPercentage:")
print(type_pct.round(2))
```

```python
print("\nQ2: Top 10 content-producing countries")
print("=" * 40)
top_countries = (df_clean[df_clean['country'] != 'Unknown']
                 ['country']
                 .str.split(',')
                 .explode()
                 .str.strip()
                 .value_counts()
                 .head(10))
print(top_countries)
```

```python
print("\nQ3: Which month sees the most content additions?")
print("=" * 40)
df_clean['month_added'] = df_clean['date_added'].dt.month_name()
df_clean['month_added'].value_counts()
```

```python
print("\nQ4: Content added per year")
print("=" * 40)
yearly = df_clean.groupby('year_added').size().reset_index(name='count')
print(yearly.sort_values('year_added'))
```

```python
print("\nQ5: Which year has the highest number of releases?")
print("=" * 40)
df_clean['release_year'].value_counts().head(10)
# OR
# df_clean.groupby('release_year').size().sort_values(ascending=False).head(10)
```

- Movies are significantly more numerous than TV Shows, indicating that Netflix focuses heavily on movie content.
